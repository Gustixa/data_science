```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from factor_analyzer.factor_analyzer import calculate_kmo, calculate_bartlett_sphericity
# Seleccionar solo las variables numéricas
numeric_df = df.select_dtypes(include=[np.number])

# KMO Test
kmo_all, kmo_model = calculate_kmo(numeric_df)
print('KMO Test:', kmo_model)

# Bartlett Test
bartlett_test, p_value = calculate_bartlett_sphericity(numeric_df)
print('Bartlett Test:', bartlett_test, 'p-value:', p_value)
```

```python
# Estandarizar los datos
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_df)

# Realizar PCA para retener el 95% de la varianza
pca_95 = PCA(n_components=0.95)
pca_95_result = pca_95.fit_transform(scaled_data)

# Explicar la varianza de cada componente
explained_variance_95 = pca_95.explained_variance_ratio_
print('Explained variance ratio (95% varianza):', explained_variance_95)

# Componentes principales
components_95 = pca_95.components_
print('Principal components (95% varianza):', components_95)

# Realizar PCA para visualización en 2D
pca_2 = PCA(n_components=2)
pca_2_result = pca_2.fit_transform(scaled_data)

# Crear el DataFrame de resultados
pca_df = pd.DataFrame(data=pca_2_result, columns=['PCA1', 'PCA2'])

# Graficar los resultados
plt.figure(figsize=(10, 6))
plt.scatter(pca_df['PCA1'], pca_df['PCA2'], alpha=0.5)
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('PCA - Representación en 2D')
plt.grid(True)
plt.show()
```
